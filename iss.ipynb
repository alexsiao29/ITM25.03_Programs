```python
import requests
import json
import time
import numpy as np
import matplotlib.pyplot as plt

def getting_data():
    req = requests.get("http://api.open-notify.org/iss-now.json")
    obj = json.loads(req.content)
    longitude=float(obj['iss_position']['longitude'])
    latitude=float(obj['iss_position']['latitude'])
    return longitude, latitude

longi=[]
lati=[]

for num in range(100):
    longitude, latitude=getting_data()
    longi.append(longitude)
    lati.append(latitude)
    time.sleep(10)

independent=np.array(lati)
dependent=np.array(longi)
    
plt.scatter(independent,dependent)
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Tracking International Space Station")
plt.show
```
